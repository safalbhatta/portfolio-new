## 5. Convolutional Neural Networks (CNN)

Build and train a CNN model using TensorFlow or PyTorch.

```python
import numpy as np
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers

# Build a simple CNN model
model = keras.Sequential([
    layers.Conv2D(32, (3, 3), activation='relu', input_shape=(28, 28, 1)),
    layers.MaxPooling2D((2, 2)),
    layers.Conv2D(64, (3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),
    layers.Conv2D(64, (3, 3), activation='relu'),
    layers.Flatten(),
    layers.Dense(64, activation='relu'),
    layers.Dropout(0.5),
    layers.Dense(10, activation='softmax')
])

# Compile the model
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

# Display model architecture
model.summary()

print("\nCNN Model created successfully!")
print("This model can be trained on datasets like MNIST for image classification.")
```

## 10. Exploratory Data Analysis (EDA)

Perform EDA on a dataset to extract insights and visualize data.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Create a sample dataset
np.random.seed(42)
eda_data = {
    'Age': np.random.randint(18, 65, 100),
    'Income': np.random.randint(20000, 150000, 100),
    'Experience': np.random.randint(0, 40, 100),
    'Education': np.random.choice(['High School', 'Bachelor', 'Master', 'PhD'], 100)
}

df_eda = pd.DataFrame(eda_data)

print("Exploratory Data Analysis (EDA):")
print("\n1. Dataset Overview:")
print(df_eda.head(10))

print("\n2. Dataset Info:")
print(f"Shape: {df_eda.shape}")
print(f"Missing values:\n{df_eda.isnull().sum()}")

print("\n3. Statistical Summary:")
print(df_eda.describe())

print("\n4. Value Counts (Education):")
print(df_eda['Education'].value_counts())

print("\n5. Correlation Analysis:")
print(df_eda[['Age', 'Income', 'Experience']].corr())

# Create visualizations
fig, axes = plt.subplots(2, 2, figsize=(12, 10))

# Age distribution
axes[0, 0].hist(df_eda['Age'], bins=20, color='skyblue', edgecolor='black')
axes[0, 0].set_title('Age Distribution')
axes[0, 0].set_xlabel('Age')
axes[0, 0].set_ylabel('Frequency')

# Income distribution
axes[0, 1].hist(df_eda['Income'], bins=20, color='lightcoral', edgecolor='black')
axes[0, 1].set_title('Income Distribution')
axes[0, 1].set_xlabel('Income')
axes[0, 1].set_ylabel('Frequency')

# Age vs Income
axes[1, 0].scatter(df_eda['Age'], df_eda['Income'], alpha=0.6, color='green')
axes[1, 0].set_title('Age vs Income')
axes[1, 0].set_xlabel('Age')
axes[1, 0].set_ylabel('Income')

# Education distribution
education_counts = df_eda['Education'].value_counts()
axes[1, 1].bar(education_counts.index, education_counts.values, color='orange', edgecolor='black')
axes[1, 1].set_title('Education Distribution')
axes[1, 1].set_ylabel('Count')
axes[1, 1].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()

print("\nEDA completed successfully!")
```

## 12. SQLite Integration

Integrate SQLite with Python for database operations.

```python
import sqlite3
import pandas as pd

# Create an in-memory SQLite database
conn = sqlite3.connect(':memory:')
cursor = conn.cursor()

# Create a table
cursor.execute('''
    CREATE TABLE students (
        id INTEGER PRIMARY KEY,
        name TEXT NOT NULL,
        age INTEGER,
        grade TEXT,
        gpa REAL
    )
''')

# Insert data
students_data = [
    (1, 'Alice', 20, 'A', 3.9),
    (2, 'Bob', 21, 'B', 3.5),
    (3, 'Charlie', 20, 'A', 3.8),
    (4, 'Diana', 22, 'A', 3.95),
    (5, 'Eve', 21, 'B', 3.6)
]

cursor.executemany('INSERT INTO students VALUES (?, ?, ?, ?, ?)', students_data)
conn.commit()

print("SQLite Integration Example:")
print("\n1. Retrieve all students:")
df_students = pd.read_sql_query('SELECT * FROM students', conn)
print(df_students)

print("\n2. Find students with grade A:")
df_grade_a = pd.read_sql_query('SELECT * FROM students WHERE grade = "A"', conn)
print(df_grade_a)

print("\n3. Calculate average GPA:")
cursor.execute('SELECT AVG(gpa) as avg_gpa FROM students')
avg_gpa = cursor.fetchone()[0]
print(f"Average GPA: {avg_gpa:.2f}")

print("\n4. Count students by grade:")
df_count = pd.read_sql_query('SELECT grade, COUNT(*) as count FROM students GROUP BY grade', conn)
print(df_count)

print("\n5. Update student data:")
cursor.execute('UPDATE students SET gpa = 3.92 WHERE name = "Alice"')
conn.commit()
print("Alice's GPA updated to 3.92")

# Close the connection
conn.close()
print("\nSQLite database operations completed!")
```

```python
# Common Data Structures Implementation

print("=" * 50)
print("COMMON DATA STRUCTURES")
print("=" * 50)

# 1. Stack
print("\n1. Stack Implementation:")
class Stack:
    def __init__(self):
        self.items = []
    
    def push(self, item):
        self.items.append(item)
    
    def pop(self):
        return self.items.pop() if not self.is_empty() else None
    
    def is_empty(self):
        return len(self.items) == 0
    
    def peek(self):
        return self.items[-1] if not self.is_empty() else None

stack = Stack()
stack.push(10)
stack.push(20)
stack.push(30)
print(f"Stack: {stack.items}")
print(f"Pop: {stack.pop()}")

# 2. Queue
print("\n2. Queue Implementation:")
from collections import deque

queue = deque()
queue.append(1)
queue.append(2)
queue.append(3)
print(f"Queue: {list(queue)}")
print(f"Dequeue: {queue.popleft()}")

# 3. Linked List
print("\n3. Linked List Implementation:")
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class LinkedList:
    def __init__(self):
        self.head = None
    
    def append(self, data):
        new_node = Node(data)
        if not self.head:
            self.head = new_node
            return
        current = self.head
        while current.next:
            current = current.next
        current.next = new_node
    
    def display(self):
        elements = []
        current = self.head
        while current:
            elements.append(current.data)
            current = current.next
        return elements

ll = LinkedList()
ll.append(1)
ll.append(2)
ll.append(3)
print(f"Linked List: {ll.display()}")

# 4. Binary Search Tree
print("\n4. Binary Search Tree:")
class TreeNode:
    def __init__(self, value):
        self.value = value
        self.left = None
        self.right = None

class BinarySearchTree:
    def __init__(self):
        self.root = None
    
    def insert(self, value):
        if not self.root:
            self.root = TreeNode(value)
        else:
            self._insert_recursive(self.root, value)
    
    def _insert_recursive(self, node, value):
        if value < node.value:
            if node.left is None:
                node.left = TreeNode(value)
            else:
                self._insert_recursive(node.left, value)
        else:
            if node.right is None:
                node.right = TreeNode(value)
            else:
                self._insert_recursive(node.right, value)
    
    def inorder(self, node=None):
        if node is None:
            node = self.root
        result = []
        if node:
            result.extend(self.inorder(node.left))
            result.append(node.value)
            result.extend(self.inorder(node.right))
        return result

bst = BinarySearchTree()
for value in [5, 3, 7, 2, 4, 6, 8]:
    bst.insert(value)
print(f"BST (Inorder): {bst.inorder()}")

# 5. Hash Table (Dictionary)
print("\n5. Hash Table (Dictionary):")
hash_table = {}
hash_table['name'] = 'Alice'
hash_table['age'] = 25
hash_table['email'] = 'alice@example.com'
print(f"Hash Table: {hash_table}")

# 6. Heap
print("\n6. Heap:")
import heapq
heap = [5, 3, 8, 1, 9, 2]
heapq.heapify(heap)
print(f"Min Heap: {heap}")
print(f"Pop min: {heapq.heappop(heap)}")

print("\nData Structures implementation completed!")
```

## 19. Algorithm Examples

Demonstrate algorithms like sorting, searching, and dynamic programming.

```python
# Algorithm Examples

print("=" * 50)
print("ALGORITHM EXAMPLES")
print("=" * 50)

# 1. Sorting Algorithms
print("\n1. Sorting Algorithms:")

def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        for j in range(0, n - i - 1):
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
    return arr

def quick_sort(arr):
    if len(arr) <= 1:
        return arr
    pivot = arr[len(arr) // 2]
    left = [x for x in arr if x < pivot]
    middle = [x for x in arr if x == pivot]
    right = [x for x in arr if x > pivot]
    return quick_sort(left) + middle + quick_sort(right)

def merge_sort(arr):
    if len(arr) <= 1:
        return arr
    mid = len(arr) // 2
    left = merge_sort(arr[:mid])
    right = merge_sort(arr[mid:])
    return merge(left, right)

def merge(left, right):
    result = []
    i = j = 0
    while i < len(left) and j < len(right):
        if left[i] <= right[j]:
            result.append(left[i])
            i += 1
        else:
            result.append(right[j])
            j += 1
    result.extend(left[i:])
    result.extend(right[j:])
    return result

arr = [64, 34, 25, 12, 22, 11, 90]
print(f"Original: {arr}")
print(f"Bubble Sort: {bubble_sort(arr.copy())}")
print(f"Quick Sort: {quick_sort(arr.copy())}")
print(f"Merge Sort: {merge_sort(arr.copy())}")

# 2. Searching Algorithms
print("\n2. Searching Algorithms:")

def linear_search(arr, target):
    for i, item in enumerate(arr):
        if item == target:
            return i
    return -1

def binary_search(arr, target):
    left, right = 0, len(arr) - 1
    while left <= right:
        mid = (left + right) // 2
        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            left = mid + 1
        else:
            right = mid - 1
    return -1

arr_sorted = sorted(arr)
target = 25
print(f"Array: {arr_sorted}")
print(f"Linear Search for {target}: {linear_search(arr_sorted, target)}")
print(f"Binary Search for {target}: {binary_search(arr_sorted, target)}")

# 3. Dynamic Programming
print("\n3. Dynamic Programming Examples:")

def fibonacci(n):
    if n <= 1:
        return n
    dp = [0] * (n + 1)
    dp[1] = 1
    for i in range(2, n + 1):
        dp[i] = dp[i - 1] + dp[i - 2]
    return dp[n]

print(f"Fibonacci(10): {fibonacci(10)}")

# 4. Recursion
print("\n4. Recursion Examples:")

def factorial(n):
    if n <= 1:
        return 1
    return n * factorial(n - 1)

print(f"Factorial(5): {factorial(5)}")

# 5. Graph Traversal
print("\n5. Graph Traversal:")

def dfs(graph, start, visited=None):
    if visited is None:
        visited = set()
    visited.add(start)
    result = [start]
    for neighbor in graph[start]:
        if neighbor not in visited:
            result.extend(dfs(graph, neighbor, visited))
    return result

graph = {
    'A': ['B', 'C'],
    'B': ['A', 'D'],
    'C': ['A', 'D'],
    'D': ['B', 'C']
}

print(f"DFS from A: {dfs(graph, 'A')}")

# 6. Two Pointer Technique
print("\n6. Two Pointer Technique:")

def two_sum(arr, target):
    left, right = 0, len(arr) - 1
    arr_sorted = sorted(arr)
    while left < right:
        current_sum = arr_sorted[left] + arr_sorted[right]
        if current_sum == target:
            return [arr_sorted[left], arr_sorted[right]]
        elif current_sum < target:
            left += 1
        else:
            right -= 1
    return None

arr = [2, 7, 11, 15]
target = 9
print(f"Two Sum for {target} in {arr}: {two_sum(arr, target)}")

print("\nAlgorithm examples completed!")
```
